```python
import cv2
import easyocr
from ultralytics import YOLO

# Load model
model = YOLO("runs/detect/train2/weights/best.pt")

# Load image
img = cv2.imread("D:/JesDocs/JesDocs/Emb_machine_vision/YOLO_dataset/images/train/Cars0.png")

# Run detection
results = model(img)

# Initialize OCR
reader = easyocr.Reader(['en'])

for r in results:
    for box in r.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)

        # Crop plate
        plate = img[y1:y2, x1:x2]

        # OCR
        text = reader.readtext(plate)

        print("Detected Plate:", text)
```

```python
cleaned_text = text[0][1].replace("O", "0")
print("Cleaned Plate:", cleaned_text)
```

```python
from collections import Counter

buffer = []

# Simulate multiple frames
for i in range(5):
    result = reader.readtext(plate)
    buffer.append(result[0][1])

final_plate = Counter(buffer).most_common(1)[0][0]

print("Final Plate:", final_plate)
```
